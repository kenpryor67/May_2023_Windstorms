```python
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import numpy.ma as ma
import h5py
import cartopy.crs as crs
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
import matplotlib.cm as cm

# Open file.
FILE_NAME = '3B-HHR-L.MS.MRG.3IMERG.20230509-S060000-E062959.0360.V06C.HDF5'
DATAFIELD_NAME = '/Grid/precipitationCal'
LATITUDE = '/Grid/lat'
LONGITUDE = '/Grid/lon'

with h5py.File(FILE_NAME, mode='r') as f:
    # List available datasets.
    print(f.keys())

    # Read dataset.
    dset = f[DATAFIELD_NAME]
    cal_precip_0600 = dset[0,:,:]
    print(cal_precip_0600,cal_precip_0600.shape)
    latitude_0600 = f[LATITUDE]
    lat_0600 = latitude_0600[:]
    print(lat_0600,lat_0600.shape)
    longitude_0600 = f[LONGITUDE]
    lon_0600 = longitude_0600[:]
    print(lon_0600,lon_0600.shape)

lat_imerg_0600,lon_imerg_0600 = np.meshgrid(lat_0600,lon_0600)
print(lat_imerg_0600,lat_imerg_0600.shape)
print(lon_imerg_0600,lon_imerg_0600.shape)
cal_precip_max_0600 = np.amax(cal_precip_0600)
cal_precip_min_0600 = np.amin(cal_precip_0600)
print("precip max =", cal_precip_max_0600)
print("precip min =", cal_precip_min_0600)
del_precip_0600 = np.gradient(cal_precip_0600)
del_precip_x_0600 = del_precip_0600[0]
del_precip_y_0600 = del_precip_0600[1]
print("Precip Gradient X = ", del_precip_x_0600)
print("Precip Gradient Y = ", del_precip_y_0600)
mag_0600 = np.sqrt(del_precip_0600[0]**2 + del_precip_0600[1]**2)
print("Precip Gradient = ", mag_0600)
cal_precip_0600[cal_precip_0600>19] = 7.5
mag_0600[mag_0600>6] = 2.5

FILE_NAME = '3B-HHR-L.MS.MRG.3IMERG.20230509-S063000-E065959.0390.V06C.HDF5'
DATAFIELD_NAME = '/Grid/precipitationCal'
LATITUDE = '/Grid/lat'
LONGITUDE = '/Grid/lon'

with h5py.File(FILE_NAME, mode='r') as f:
    # List available datasets.
    print(f.keys())

    # Read dataset.
    dset = f[DATAFIELD_NAME]
    cal_precip_0630 = dset[0,:,:]
    print(cal_precip_0630,cal_precip_0630.shape)
    latitude_0630 = f[LATITUDE]
    lat_0630 = latitude_0630[:]
    print(lat_0630,lat_0630.shape)
    longitude_0630 = f[LONGITUDE]
    lon_0630 = longitude_0630[:]
    print(lon_0630,lon_0630.shape)

lat_imerg_0630,lon_imerg_0630 = np.meshgrid(lat_0630,lon_0630)
print(lat_imerg_0630,lat_imerg_0630.shape)
print(lon_imerg_0630,lon_imerg_0630.shape)
cal_precip_0630_max = np.amax(cal_precip_0630)
cal_precip_0630_min = np.amin(cal_precip_0630)
print("precip max =", cal_precip_0630_max)
print("precip min =", cal_precip_0630_min)
del_precip_0630 = np.gradient(cal_precip_0630)
del_precip__0630_x = del_precip_0630[0]
del_precip__0630_y = del_precip_0630[1]
print("Precip Gradient X = ", del_precip__0630_x)
print("Precip Gradient Y = ", del_precip__0630_y)
mag_0630 = np.sqrt(del_precip_0630[0]**2 + del_precip_0630[1]**2)
print("Precip Gradient = ", mag_0630)
cal_precip_0630[cal_precip_0630>19] = 7.5
mag_0630[mag_0630>6] = 2.5

FILE_NAME = '3B-HHR-L.MS.MRG.3IMERG.20230509-S070000-E072959.0420.V06C.HDF5'
DATAFIELD_NAME = '/Grid/precipitationCal'
LATITUDE = '/Grid/lat'
LONGITUDE = '/Grid/lon'

with h5py.File(FILE_NAME, mode='r') as f:
    # List available datasets.
    print(f.keys())

    # Read dataset.
    dset = f[DATAFIELD_NAME]
    cal_precip_0700 = dset[0,:,:]
    print(cal_precip_0700,cal_precip_0700.shape)
    latitude_0700 = f[LATITUDE]
    lat_0700 = latitude_0700[:]
    print(lat_0700,lat_0700.shape)
    longitude_0700 = f[LONGITUDE]
    lon_0700 = longitude_0700[:]
    print(lon_0700,lon_0700.shape)

lat_imerg_0700,lon_imerg_0700 = np.meshgrid(lat_0700,lon_0700)
print(lat_imerg_0700,lat_imerg_0700.shape)
print(lon_imerg_0700,lon_imerg_0700.shape)
cal_precip_0700_max = np.amax(cal_precip_0700)
cal_precip_0700_min = np.amin(cal_precip_0700)
print("precip max =", cal_precip_0700_max)
print("precip min =", cal_precip_0700_min)
del_precip_0700 = np.gradient(cal_precip_0700)
del_precip__0700_x = del_precip_0700[0]
del_precip__0700_y = del_precip_0700[1]
print("Precip Gradient X = ", del_precip__0700_x)
print("Precip Gradient Y = ", del_precip__0700_y)
mag_0700 = np.sqrt(del_precip_0700[0]**2 + del_precip_0700[1]**2)
print("Precip Gradient = ", mag_0700)
cal_precip_0700[cal_precip_0700>27] = 10
mag_0700[mag_0700>6] = 2.5
```
